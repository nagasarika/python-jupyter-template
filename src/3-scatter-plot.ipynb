### Exercise 3: Scatter Plot with Annotations

```python
# TODO: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Set figure size
plt.figure(figsize=(6, 4))

# Create the scatter plot with larger markers and a label
plt.scatter(x, y, color='blue', marker='o', s=100, label="Data Points")

# Annotate each point with its (x, y) value
for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(10, 10), ha='center')

# Customize the plot
plt.title('Scatter Plot with Annotations')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.legend()  # Add legend

# Show the plot
plt.show()
```
